# $$\text{Transform!!}$$

```r
library(dplyr) ## useful for transforming dataframes
#https://cran.r-project.org/web/packages/dplyr/dplyr.pdf
```

```r
tran2<-dataT %>%
    select_if(is.character)
```

```r
tran1<-dataT %>%
    mutate_if(is.integer,as.factor)
```

```r
tran1<-tran1 %>%
    select_if(is.factor)
```

```r
dim(tran1)
## so 154 variables as factors thus far
```

```r
summary(tran1)
#count of the various factor levels for each variable
```

```r
summary(tran2) ## list of all none factors
## we are going to transform these to dummy variables
```

```r
install.packages("fastDummies")
```

```r
library(fastDummies) ## used to create dummy variables ...fast..
```

```r
## Transforming tran 2 varaibles into dummy variables 
datadum1=dummy_cols(tran2,remove_first_dummy=T)
## datadum1 will be used later on for modeling or comparison 
```

```r
glimpse(datadum1)
```

```r
## do we have any zero variance in datadum1?
which(apply(datadum1,2,var) ==0)
## so a few of our chacter variables had some na's attached to them
## Also, it looks like dragon, unicorns,kings,and starocean lacked variation.
## Theres are all the variables I added,to test for a quick way to find simple data error
## we can now remove them, since they are useless 
## The numbers below represent their column postion in datadum1
## If the orginal variable had NA as well, a an attriable would have been created with ..._NA
```

```r
# Removing dragon, unicorns,kings,and starocean columns 
tran2<-tran2 %>%
    select(everything(),-c(dragons, unicorns,kings,starocean))
```

```r
datadum1=dummy_cols(tran2,remove_first_dummy=T)
which(apply(datadum1,2,var) ==0)
```

```r
tran3<-tran2 %>%
    mutate_if(is.character,as.factor)
## will cbind tran3 and tran2 together, after fixing na issues in tran1
```

```r
dataNT<-cbind(tran1,tran3)
```
